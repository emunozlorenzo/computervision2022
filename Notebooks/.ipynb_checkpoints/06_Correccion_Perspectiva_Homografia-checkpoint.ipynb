# Correción de Perspectiva Homografía

__OBJETIVO__ Cambiar la perspectiva de una imagen utilizando 4 puntos inciales y 4 puntos finales

![img](../images/homografia.png)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../images/letrero.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap='gray');
```

```python
# Sacamos dimensiones de la imagen
m, n, _ = img.shape
print(f"Largo: {m} px Ancho: {n} px")
```

```python
# Sacamos los vértices iniciales de la imagen que queremos corregir en perspectiva
pts_src = np.array([[443,751],[1503,352],[439,1177],[1517,905]])
# Marcamos los vértices finales
pts_dst = np.array([[0,0],[n,0],[0,m],[n,m]])
```

```python
# h sería la matriz de la homografía
h,_ = cv2.findHomography(pts_src, pts_dst)
im2 = cv2.warpPerspective(img, h, (n,m))
```

```python
fig, axs = plt.subplots(1,2, figsize=(10, 10))
axs[0].title.set_text('ORIGINAL')
axs[1].title.set_text('HOMOGRAFÍA')
axs[0].imshow(img, cmap='gray');
axs[1].imshow(im2, cmap='gray');
```

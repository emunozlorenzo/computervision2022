# Medición de Parámetros en una imagen

Important Notes:

- OpenCV and Matplotlib expect different orders of the RGB channels
    - Matplotlib Orders: RED GREEN BLUE
    - OpenCV Orders: BLUE GREEN RED

## Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
from PIL import Image
```

## Read Data

```python
# Abrimos la imagen con openCV: Se trata de una matriz de numpy
img = cv2.imread('../images/000078.jpg')
type(img)
```

## Create Mask

```python
type(img)
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
umbral,_ = cv2.threshold(gray_img, 0, 255, cv2.THRESH_OTSU)
print('Método OTSU determina umbral en:',umbral)
mask = np.uint8((gray_img<umbral)*255)
plt.imshow(mask, cmap='gray');
```

## Detección de objetos en la máscara

```python
output = cv2.connectedComponentsWithStats(mask, 4, cv2.CV_32S)
cantidad_objetos = output[0]
labels = output[1]
stats = output[2]
print('Cantidad de Objetos detectados:',cantidad_objetos)
plt.imshow(labels);
```

```python
# Columna 4 es el numero de pixeles de cada objeto por lo tanto la primera fila es el fondo
stats
```

```python
mascara = (np.argmax(stats[:,4][1:])+1==labels)
plt.imshow(mascara);
```

## Rellenar huecos de máscara

```python
# ¿Cómo rellenar huecos?
from scipy import ndimage
new_mascara = ndimage.binary_fill_holes(mascara).astype(int)
plt.imshow(new_mascara);
```

## ConvexHull
 - Póligono convexo más pequeño que encierra el objeto
 
![img](https://miro.medium.com/max/1400/1*-oQr4F4lYcMRO4l4oCgXjg.png)

```python
for c in contours:
    hull = cv2.convexHull(c)
    puntosConvexHull = hull[:, 0, :]
    m, n = mascara8bits.shape
    ar = np.zeros((m,n))
    mascaraConvex = cv2.fillConvexPoly(ar, puntosConvexHull, 1)
    plt.imshow(mascaraConvex);
```

## Bounding Box Recto

#### BoundingBox Rotado

```python
mascara8bits =np.uint8(new_mascara+255)
for c in contours:
    rect = cv2.minAreaRect(c)
    box = cv2.boxPoints(rect)
    # convert all coordinates floating point values to int
    box = np.int0(box)
    # draw a red 'nghien' rectangle
    rectangulo1 = cv2.drawContours(mascara8bits, [box], 0, (0, 0, 255))
    plt.imshow(rectangulo1);
```

#### BoundingBox Recto

```python
mascara8bits =np.uint8(new_mascara+255)
for c in contours:
    # get the bounding rect
    x, y, w, h = cv2.boundingRect(c)
    # draw a green rectangle to visualize the bounding rect
    rectangulo2 = cv2.rectangle(mascara8bits, (x, y), (x+w, y+h), (0, 255, 0), 1)
    plt.imshow(rectangulo2);
```

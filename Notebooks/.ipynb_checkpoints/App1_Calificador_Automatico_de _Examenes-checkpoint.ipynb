#### 1. Segmentación

- 1.1 Canny: Algoritmo para Edge Detection
- 1.2 Selección objeto con mayor área
- 1.3 ConvexHull
- 1.4 Corregir la perspectiva
- 1.5 máscara

![img](../images/calificador.png)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#leemos imagen en escala de grises
img = cv2.imread('../images/foto_formato2.jpg', 0)

fig, axs = plt.subplots(1,1, figsize=(10, 10))
axs.imshow(img, cmap='gray');
```

##### 1.1 Canny

Algoritmo para Edge Detection

```python
# Segmentamos con Algoritmo de Canny
canny = cv2.Canny(img, 20,150)
# Ampliamos un poco los bordes del objeto, DILATACIÓN
kernel =np.ones((10,10), np.uint8)
bordes = cv2.dilate(canny, kernel)

# Comparamos Resultados de ambos
fig, axs = plt.subplots(1,3, figsize=(20, 20))
axs[0].title.set_text('ORIGINAL')
axs[1].title.set_text('CANNY')
axs[2].title.set_text('CANNY DILATE')
axs[0].imshow(img, cmap='gray');
axs[1].imshow(canny, cmap='gray');
axs[2].imshow(bordes, cmap='gray');
```

```python
# Generamos la máscara del objeto más grande (el de área máxima)
contours,_ = cv2.findContours(bordes,  cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
objetos = bordes.copy()
cv2.drawContours(objetos, [max(contours, key=cv2.contourArea)],-1,255,thickness=-1)

fig, axs = plt.subplots(1,1, figsize=(10, 10))
axs.imshow(objetos, cmap='gray');
```

##### 1.2 Detección del área máxima

Detección del área de interes

- Detectar área máxima

```python
# Necesitamos conseguir con solo el objeto de mayor área, ¿cómo lo identificamos? connectedComponentWithStats
output = cv2.connectedComponentsWithStats(objetos, 4, cv2.CV_32S)
# Cantidad de objetos
cantidad_objetos = output[0]
print('Cantidad de Objetos detectados:',cantidad_objetos)
# Etiquetas
labels = output[1]
# Estadísticas (columna 4 contiene el área)
stats = output[2] 

# Máscara del área máxima (después del fondo) (Cúal es la etiqueta del objeto que tiene mayor áre)
mascara = np.uint8(255*(np.argmax(stats[:,4][1:])+1==labels))
                   
fig, axs = plt.subplots(1,1, figsize=(10, 10))
axs.imshow(mascara, cmap='gray');
```

##### 1.3 ConvexHull

- ConvexHull: nos ayuda a identificar mejor los vértices del polígono

```python
# Hallar el ConvexHull para corregir regiones que tienen picos y sea más fácil hallar los vértices del polígono
contours,_ = cv2.findContours(mascara,  cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
cnt = contours[0]

for c in contours:
    hull = cv2.convexHull(c)
    puntosConvexHull = hull[:, 0, :]
    m, n = mascara.shape
    ar = np.zeros((m,n))
    mascaraConvex = np.uint8(cv2.fillConvexPoly(ar, puntosConvexHull, 1))
                   
fig, axs = plt.subplots(1,1, figsize=(10, 10))
axs.imshow(mascaraConvex, cmap='gray');
```

##### 1.4 Corregir la perspectiva

- Para corregir la perspectiva se necesitan 4 puntos iniciales y 4 puntos finales

```python
# Forma de detectar los 4 vértices de la máscara, Vértices Iniciales
vertices = cv2.goodFeaturesToTrack(mascaraConvex, 4, 0.01, 20)
x = vertices[:, 0, 0]
y = vertices[:, 0, 1]
vertices = vertices[:,0,:] # pasa de 4,1,2 a 4,2 para después hacer la homografía
vertices = np.int64(vertices)

# Forma de detectar los vértices finales (serán los vertices de la imagen completa)
xo = np.sort(x)
yo = np.sort(y)
xn = np.zeros((1,4))
yn = np.zeros((1,4))
xn = (x==xo[2])*n+(x==xo[3])*n
yn = (y==yo[2])*m+(y==yo[3])*m
# Vertices Finales
verticesN = np.zeros((4,2))
verticesN[:,0] = xn
verticesN[:,1] = yn
verticesN = np.int64(verticesN)
```

```python
# h sería la matriz de la homografía
h,_ = cv2.findHomography(vertices, verticesN)
im2 = cv2.warpPerspective(img, h, (n,m))
fig, axs = plt.subplots(1,2, figsize=(10, 10))
axs[0].title.set_text('ORIGINAL')
axs[1].title.set_text('HOMOGRAFÍA')
axs[0].imshow(img, cmap='gray');
axs[1].imshow(im2, cmap='gray');
```

#### 2. División Vertical del Formato y Cálculo de Respuestas

```python
# Región de interés desde el 25% de la anchura al 85% de la anchura de la imagen de esta forma nos independizamos del tamaño de la imagen original para encontrar la ROI
roi = im2[:,np.uint64(0.25*n):np.uint64(0.85*n)]
fig, axs = plt.subplots(1,1, figsize=(10, 10))
axs.imshow(roi, cmap='gray');
```

```python
fig, axs = plt.subplots(1,1, figsize=(10, 10))
axs.imshow(roi[:np.uint64(m/26),:], cmap='gray');
```

```python
# Lista de Opciones
opciones = ['A','B','C','D','E','NS/NC']
respuestas = []
resta = []
for i in range(0,26):
    pregunta = roi[np.uint64(i*(m/26)):np.uint64((i+1)*(m/26)),:]
    sumI = []
    for j in range(0,5):
        _, col = pregunta.shape
        sumI.append(np.sum(pregunta[:,np.uint64(j*(col/5)):np.uint64((j+1)*(col/5))]))

    # Método para detectar las preguntas no contestadas
    vmin = np.ones((1,5))*np.min(sumI)
    if np.linalg.norm(sumI-vmin) > 0.17*col*m: # 1000000
        sumI.append(float('inf'))
    else:
        sumI.append(-1)        
    respuestas.append(opciones[np.argmin(sumI)])

    fig, axs = plt.subplots(1,1, figsize=(10, 10))
    axs.title.set_text(f"Pregunta: {i+1}. Respuesta Calculada: {opciones[np.argmin(sumI)]}. Intensidades de Cada Respuesta: {sumI}")
    axs.imshow(roi[np.uint64(i*(m/26)):np.uint64((i+1)*(m/26)),:], cmap='gray');
```

#### 3. Encapsulamos Código

```python
# Lo ponemos en forma de función
def obtenerRespuestas(img):
    # Segmentamos con Algoritmo de Canny
    canny = cv2.Canny(img, 20,150)
    # Ampliamos un poco los bordes del objeto, DILATACIÓN
    kernel =np.ones((10,10), np.uint8)
    bordes = cv2.dilate(canny, kernel)
    
    # Generamos la máscara del objeto más grande (el de área máxima)
    contours,_ = cv2.findContours(bordes,  cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    objetos = bordes.copy()
    cv2.drawContours(objetos, [max(contours, key=cv2.contourArea)],-1,255,thickness=-1)
    
    # Necesitamos conseguir con solo el objeto de mayor área, ¿cómo lo identificamos? connectedComponentWithStats
    output = cv2.connectedComponentsWithStats(objetos, 4, cv2.CV_32S)
    # Cantidad de objetos
    cantidad_objetos = output[0]
    # Etiquetas
    labels = output[1]
    # Estadísticas (columna 4 contiene el área)
    stats = output[2] 

    # Máscara del área máxima (después del fondo) (Cúal es la etiqueta del objeto que tiene mayor áre)
    mascara = np.uint8(255*(np.argmax(stats[:,4][1:])+1==labels))
    
    # Hallar el ConvexHull para corregir regiones que tienen picos y sea más fácil hallar los vértices del polígono
    contours,_ = cv2.findContours(mascara,  cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    cnt = contours[0]

    for c in contours:
        hull = cv2.convexHull(c)
        puntosConvexHull = hull[:, 0, :]
        m, n = mascara.shape
        ar = np.zeros((m,n))
        mascaraConvex = np.uint8(cv2.fillConvexPoly(ar, puntosConvexHull, 1))
    
    # Forma de detectar los 4 vértices de la máscara, Vértices Iniciales
    vertices = cv2.goodFeaturesToTrack(mascaraConvex, 4, 0.01, 20)
    x = vertices[:, 0, 0]
    y = vertices[:, 0, 1]
    vertices = vertices[:,0,:] # pasa de 4,1,2 a 4,2 para después hacer la homografía
    vertices = np.int64(vertices)

    # Forma de detectar los vértices finales (serán los vertices de la imagen completa)
    xo = np.sort(x)
    yo = np.sort(y)
    xn = np.zeros((1,4))
    yn = np.zeros((1,4))
    xn = (x==xo[2])*n+(x==xo[3])*n
    yn = (y==yo[2])*m+(y==yo[3])*m
    # Vertices Finales
    verticesN = np.zeros((4,2))
    verticesN[:,0] = xn
    verticesN[:,1] = yn
    verticesN = np.int64(verticesN)
    
    # h sería la matriz de la homografía
    h,_ = cv2.findHomography(vertices, verticesN)
    im2 = cv2.warpPerspective(img, h, (n,m))
    
    
    # Lista de Opciones
    opciones = ['A','B','C','D','E','NS/NC']
    respuestas = []
    resta = []
    for i in range(0,26):
        pregunta = roi[np.uint64(i*(m/26)):np.uint64((i+1)*(m/26)),:]
        sumI = []
        for j in range(0,5):
            _, col = pregunta.shape
            sumI.append(np.sum(pregunta[:,np.uint64(j*(col/5)):np.uint64((j+1)*(col/5))]))

        # Método para detectar las preguntas no contestadas
        vmin = np.ones((1,5))*np.min(sumI)
        if np.linalg.norm(sumI-vmin) > 0.17*col*m: # 1000000
            sumI.append(float('inf'))
        else:
            sumI.append(-1)        
        respuestas.append(opciones[np.argmin(sumI)])
    return respuestas
```

```python
#leemos imagen en escala de grises
img = cv2.imread('../images/foto_formato2.jpg', 0)

fig, axs = plt.subplots(1,1, figsize=(10, 10))
axs.imshow(img, cmap='gray');
```

```python
obtenerRespuestas(img)
```

#### 4. Cálculo de Nota Final

```python
respuestasCorrectas = ['B','B','D','E','B','A','A','A','C','B','E','C','C','A','E','B','C','B','D','D','B','D','A','A','C','C']
calificacion = 10*np.sum(np.array(respuestasCorrectas) == np.array(obtenerRespuestas(img)))/26
```

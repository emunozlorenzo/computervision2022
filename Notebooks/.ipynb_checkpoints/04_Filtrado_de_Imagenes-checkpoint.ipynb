# Filtrado de Im√°genes

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../images/bananos.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img);
```

## 1. Filtro Promedio

```python
# Creamos una matriz de 1s
kernel_3x3 = np.ones((3,3), np.float32)/(3*3)
```

```python
kernel_3x3
```

```python
output = cv2.filter2D(img, -1, kernel_3x3)
plt.imshow(output);
```

```python
# Kernel 5x5
kernel_5x5 = np.ones((5,5), np.float32)/(5*5)
output = cv2.filter2D(img, -1, kernel_5x5)
plt.imshow(output);
```

```python
# Kernel 31x31
kernel_31x31 = np.ones((31,31), np.float32)/(31*31)
output = cv2.filter2D(img, -1, kernel_31x31)
plt.imshow(output);
```

## 2. Filtro Gaussiano

```python
# Gaussiano Desviaci√≥n 3x3
output = cv2.GaussianBlur(img,(3,3), 0)
plt.imshow(output);
```

```python
# Gaussiano Desviaci√≥n 11x11
output = cv2.GaussianBlur(img,(11,11), 0)
plt.imshow(output);
```

```python
# Gaussiano Desviaci√≥n 21x21
output = cv2.GaussianBlur(img,(21,21), 0)
plt.imshow(output);
```
